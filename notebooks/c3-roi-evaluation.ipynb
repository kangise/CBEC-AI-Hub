## 1. 准备工作

```python
import pandas as pd
import numpy as np

print('环境准备完成 ✅')
```

## 2. 定义你的 AI 工具投入

修改下面的数据为你的实际情况。

```python
# === 修改这里为你的实际数据 ===

ai_tools = {
    'ChatGPT Plus': {
        'monthly_cost': 20,  # 美元/月
        'tasks': [
            {'name': 'Listing 生成', 'hours_saved_per_month': 8, 'hourly_rate': 25},
            {'name': '广告文案', 'hours_saved_per_month': 4, 'hourly_rate': 25},
            {'name': '客服模板', 'hours_saved_per_month': 3, 'hourly_rate': 20},
            {'name': '数据分析', 'hours_saved_per_month': 5, 'hourly_rate': 30},
            {'name': '社交媒体文案', 'hours_saved_per_month': 6, 'hourly_rate': 25},
        ]
    },
    'Helium 10': {
        'monthly_cost': 79,
        'tasks': [
            {'name': '关键词研究', 'hours_saved_per_month': 10, 'hourly_rate': 30},
            {'name': '竞品分析', 'hours_saved_per_month': 6, 'hourly_rate': 30},
            {'name': 'Listing 审计', 'hours_saved_per_month': 3, 'hourly_rate': 25},
        ]
    },
    'Canva Pro': {
        'monthly_cost': 13,
        'tasks': [
            {'name': '产品图片设计', 'hours_saved_per_month': 5, 'hourly_rate': 35},
            {'name': '社交媒体图片', 'hours_saved_per_month': 4, 'hourly_rate': 25},
            {'name': '广告素材', 'hours_saved_per_month': 3, 'hourly_rate': 30},
        ]
    }
}

print(f'已配置 {len(ai_tools)} 个 AI 工具')
for name, tool in ai_tools.items():
    print(f'  - {name}: ${tool["monthly_cost"]}/月, {len(tool["tasks"])} 个应用场景')
```

## 3. 计算每个工具的 ROI

```python
results = []

for tool_name, tool in ai_tools.items():
    total_hours_saved = 0
    total_value_saved = 0
    task_details = []
    
    for task in tool['tasks']:
        value = task['hours_saved_per_month'] * task['hourly_rate']
        total_hours_saved += task['hours_saved_per_month']
        total_value_saved += value
        task_details.append({
            'Task': task['name'],
            'Hours Saved/Month': task['hours_saved_per_month'],
            'Hourly Rate': f"${task['hourly_rate']}",
            'Value Saved/Month': f"${value}"
        })
    
    monthly_cost = tool['monthly_cost']
    monthly_roi = ((total_value_saved - monthly_cost) / monthly_cost) * 100
    annual_net = (total_value_saved - monthly_cost) * 12
    
    results.append({
        'Tool': tool_name,
        'Monthly Cost': monthly_cost,
        'Hours Saved/Month': total_hours_saved,
        'Value Saved/Month': total_value_saved,
        'Monthly Net': total_value_saved - monthly_cost,
        'Monthly ROI': monthly_roi,
        'Annual Net Savings': annual_net
    })
    
    print(f'\n=== {tool_name} ===')
    print(pd.DataFrame(task_details).to_string(index=False))
    print(f'\n月成本: ${monthly_cost} | 月节省价值: ${total_value_saved} | 月净收益: ${total_value_saved - monthly_cost}')
    print(f'月 ROI: {monthly_roi:.0f}% | 年净收益: ${annual_net:,.0f}')
```

## 4. 汇总对比

```python
summary = pd.DataFrame(results)
summary['Monthly Cost'] = summary['Monthly Cost'].apply(lambda x: f'${x}')
summary['Value Saved/Month'] = summary['Value Saved/Month'].apply(lambda x: f'${x}')
summary['Monthly Net'] = summary['Monthly Net'].apply(lambda x: f'${x}')
summary['Monthly ROI'] = summary['Monthly ROI'].apply(lambda x: f'{x:.0f}%')
summary['Annual Net Savings'] = summary['Annual Net Savings'].apply(lambda x: f'${x:,.0f}')

print('=== AI 工具 ROI 汇总 ===')
print(summary.to_string(index=False))

total_monthly_cost = sum(t['monthly_cost'] for t in ai_tools.values())
total_monthly_value = sum(r['Value Saved/Month'] for r in [{'Value Saved/Month': sum(t['hours_saved_per_month'] * t['hourly_rate'] for t in tool['tasks'])} for tool in ai_tools.values()])

print(f'\n--- 总计 ---')
print(f'AI 工具总月成本: ${total_monthly_cost}')
total_val = sum(sum(t['hours_saved_per_month'] * t['hourly_rate'] for t in tool['tasks']) for tool in ai_tools.values())
print(f'总月节省价值: ${total_val}')
print(f'总月净收益: ${total_val - total_monthly_cost}')
print(f'总年净收益: ${(total_val - total_monthly_cost) * 12:,.0f}')
print(f'综合 ROI: {((total_val - total_monthly_cost) / total_monthly_cost) * 100:.0f}%')
```

## 5. 生成 AI 分析 Prompt

把你的 ROI 数据整理成 Prompt，让 AI 帮你优化工具组合。

```python
tool_summary = '\n'.join([
    f"- {r['Tool']}: 月成本 {r['Monthly Cost']}, 月节省 {r['Value Saved/Month']}, ROI {r['Monthly ROI']}"
    for r in results
])

prompt = f"""你是一个 AI 工具投资回报分析专家。

以下是我当前使用的 AI 工具及其 ROI 数据：

{tool_summary}

总月成本: ${total_monthly_cost}
总月节省: ${total_val}
综合 ROI: {((total_val - total_monthly_cost) / total_monthly_cost) * 100:.0f}%

请分析：
1. 哪个工具的 ROI 最高？为什么？
2. 有没有工具可以替换或取消？
3. 还有哪些 AI 工具值得考虑加入？（基于我的使用场景）
4. 如何进一步提升 AI 工具的使用效率？
5. 给团队的 AI 工具预算建议（月度/年度）
"""

print('=== 复制以下 Prompt 到 ChatGPT/Claude ===')
print(prompt)
```
